```python
import os
import json
import random

# 定义数据路径和可用受试者
data_path = "/root/RingTool/data/rings/"
config_dir = "/root/RingTool/config/" # 不仅仅在supervised目录下检索，在更上面一层检索

# 确保配置目录存在
if not os.path.exists(config_dir):
    print(f"Config directory {config_dir} does not exist!")
    exit(1)

# 遍历 supervised 文件夹下的所有 JSON 文件
config_files = [os.path.join(root, f) for root, _, files in os.walk(config_dir) for f in files if f.endswith(".json")]

print(f"Found {len(config_files)} JSON config files in {config_dir}")

# 为 5 折交叉验证生成划分
random.seed(42)  # 设置随机种子以确保可重现
available_subjects = [f"{i:05d}" for i in range(1, 33)]  # 生成 00001 到 00032 的受试者列表
random.shuffle(available_subjects)  # 随机打乱受试者列表

# 目前有32个受试者，每个折7个受试者，77666
five_fold = {
    "Fold-1":  list(available_subjects[:7]),
    "Fold-2":  list(available_subjects[7:14]),
    "Fold-3":  list(available_subjects[14:20]),
    "Fold-4":  list(available_subjects[20:26]),
    "Fold-5":  list(available_subjects[26:33])
}
# 
# 遍历每个 JSON 文件并更新 split 配置
for config_path in config_files:
    try:
        # 加载配置
        with open(config_path, "r") as f:
            config = json.load(f)

        # 更新 split
        config["split"]["train"] = list(available_subjects[:20])  # 前20个作为训练集
        config["split"]["valid"] = list(available_subjects[20:26])  # 接下来的6个作为验证集
        config["split"]["test"] =  list(available_subjects[26:])  # 最后6个作为测试集
        config["split"]["5-Fold"] = five_fold

        # 保存更新后的配置文件
        with open(config_path, "w") as f:
            json.dump(config, f, indent=4)

        # 打印更新信息
        print(f"\nUpdated config: {config_path}")
        print(f"Split: train={config['split']['train']}, valid={config['split']['valid']}, test={config['split']['test']}")
        print(f"5-Fold: {config['split']['5-Fold']}")
        print(f"Input type: {config['dataset']['input_type']}")
        print(f"Batch size: {config['dataset']['batch_size']}")
        print(f"Device: {config['train']['device']}")

    except Exception as e:
        print(f"Error processing {config_path}: {str(e)}")

print("\nAll configurations have been updated!")
```

```python
import os
import json
import random

# 定义数据路径和可用受试者
data_path = "/root/RingTool/data/rings/"
config_dir = "/root/RingTool/config/" # 不仅仅在supervised目录下检索，在更上面一层检索

# 确保配置目录存在
if not os.path.exists(config_dir):
    print(f"Config directory {config_dir} does not exist!")
    exit(1)

# 遍历 supervised 文件夹下的所有 JSON 文件
config_files = [os.path.join(root, f) for root, _, files in os.walk(config_dir) for f in files if f.endswith(".json")]

print(f"Found {len(config_files)} JSON config files in {config_dir}")

# 遍历每个 JSON 文件并更新 split 配置
for config_path in config_files:
    try:
        # 如果path中包含mamba字段，才进行下面的处理
        if "mamba" not in config_path:
            continue

        # 加载配置
        with open(config_path, "r") as f:
            config = json.load(f)

        # 更新
        config["dataset"]["batch_size"] = 8
        config["test"]["batch_size"] = 8

        # 保存更新后的配置文件
        with open(config_path, "w") as f:
            json.dump(config, f, indent=4)

        # 打印更新信息
        print(f"\nUpdated config: {config_path}")
        print(f"Split: train={config['split']['train']}, valid={config['split']['valid']}, test={config['split']['test']}")
        print(f"5-Fold: {config['split']['5-Fold']}")
        print(f"Input type: {config['dataset']['input_type']}")
        print(f"Batch size: {config['dataset']['batch_size']}")
        print(f"Device: {config['train']['device']}")

    except Exception as e:
        print(f"Error processing {config_path}: {str(e)}")

print("\nAll configurations have been updated!")
```
